Detecting outlier using Z score using Z score formula for Z score

```python
import random
dataset = np.random.randint(150, size=100)
print(dataset)
```

```python
from scipy.stats.mstats_basic import stde_median
outliers = []
def detect_outliers(dataset):
  threshold = 1
  # above 3rd standard deviation consider it as an outlier
  mean = np.mean(dataset)
  std = np.std(dataset)
  for i in dataset:
    z_score = (i - mean) / std
    if np.abs(z_score) > threshold:
      outliers.append(i)
  return outliers
```

```python
detect_outliers(dataset)
```

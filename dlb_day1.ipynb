```python
# selecting input feature and output label
x = df.iloc[:,0:10].values # taking first 10 colums and all rows
y = df.iloc[:,10].values  
```

```python
# step 4: dividing into training and test dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)
```

```python
# step 5: feature scaling
# convert large values to smaller value to apply regression
min_max_scaler = preprocessing.MinMaxScaler()
x_scale = min_max_scaler.fit_transform(x_train)
```

```python
x_scale
```

```python
# classifies the data using linear model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy : ', accuracy_score(y_test, y_pred))
```

```python
# get the dataset from the sklearn package
from sklearn.datasets import load_boston
dataset = load_boston() 
```

```python
data = pd.DataFrame(dataset.data)
data.columns = dataset.feature_names
```

```python
data.head()
```

```python
data['PRICE'] = dataset.target 
```

```python
data.info()
```

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
# convert the data into matrix format so that we can apply xgb(extra gradient boost)
data_matrix = xgb.DMatrix(data=x, label=y)
```

```python
# splitting the dataset into train and test dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
xg_reg = xgb.XGBRegressor(objective='reg:linear', 
                          colsample_bytree=0.3, 
                          max_depth=5,
                          alpha=10,
                          n_estimator=10,
                          learning_rate=0.3)
```

```python
xg_reg.fit(x_train, y_train)
pred = xg_reg.predict(x_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, pred))
print('rmse: ', rmse)
```

```python
xgb.plot_tree(xg_reg)
```

```python
# num_boost_round: no of trees you build
params = {'objective':'reg:linear', 
          'colsample_bytree':0.3, 
          'max_depth':5,
          'alpha':10,
          'n_estimator':10,
          'learning_rate':0.3}
cv_results = xgb.cv(dtrain=data_matrix, 
                    params=params, 
                    nfold=5, 
                    num_boost_round=50, 
                    early_stopping_rounds=10, 
                    metrics="rmse", 
                    as_pandas=True, 
                    seed=123)
```

```python
cv_results.head()
```

```python
print((cv_results["test-rmse-mean"]).tail(1))
```

```python
xg_reg = xgb.train(params=params, dtrain=data_matrix, num_boost_round=10)
```

```python
import matplotlib.pyplot as plt
xgb.plot_tree(xg_reg,num_trees=0)
plt.rcParams['figure.figsize'] = [100, 10]
plt.show()
```

```python
xgb.plot_importance(xg_reg)
plt.rcParams['figure.figsize'] = [25,15]
plt.show()
```

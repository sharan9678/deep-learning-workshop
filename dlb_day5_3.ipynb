```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/deep_learning_bootcamp/day5/car_evaluation.csv')
```

```python
dataset.head()
```

```python
# renaming the column dataset
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class']
dataset.columns = col_names
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
# to get the count of 
for col in col_names:
  print(dataset[col].value_counts())
```

```python
# checking for any missing value in the dataset
dataset.isnull().sum()
```

```python
# separating the dataset
X = dataset.drop(['class'], axis=1)
y = dataset['class']
```

```python
# split into training and testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 20)
```

```python
X_train.dtypes
```

```python
# import category encoders
!pip install category_encoders
```

```python
import category_encoders as ce
# ce is used to scale dataset with both string and number
```

```python
# encode variable with ordinal encoding
encoder = ce.OrdinalEncoder(cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])
X_train = encoder.fit_transform(X_train)
X_test = encoder.transform(X_test)
```

```python
X_train.head()
```

```python
# traning the dataset
# import the DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_gini = DecisionTreeClassifier(criterion='gini', max_depth=20, random_state=0)
# fit the model
clf_gini.fit(X_train, y_train)
```

```python
y_pred_gini = clf_gini.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('model accuracy is: {0:0.4f}'.format(accuracy_score(y_test,y_pred_gini)))
print(confusion_matrix(y_test, y_pred_gini))
```

```python
# instantiate the DecisionTreeClassifier with criteron entropy
clf_en = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
# fit the model
clf_en.fit(X_train, y_train)
```

```python
y_pred_en = clf_en.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('model accuracy is: {0:0.4f}'.format(accuracy_score(y_test,y_pred_en)))
print(confusion_matrix(y_test, y_pred_en))
```

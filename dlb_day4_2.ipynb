```python
# import numpy and load_loadwine
import numpy as np
from sklearn.datasets import load_wine
```

```python
# load the dataset
wine_dataset = load_wine()
```

```python
# get the features and the label from the dataset
wine_dataset_features = wine_dataset.data
wine_dataset_labels = wine_dataset.target
```

```python
# printing the features and labels of the dataset
print(wine_dataset_features)
print(wine_dataset_labels)
```

```python
# split the dataset into training and test dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(wine_dataset_features, wine_dataset_labels, test_size=0.20, random_state=28)
```

```python
# load the k-nearest neighbor classifier
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)
print(knn_classifier)
```

```python
# fitting the classifier on the training dataset
knn_classifier.fit(x_train, y_train)
```

```python
print(np.shape(x_test))
```

```python
# making the prediction with knn classifier, comparing it with ground truth and getting the score value
knn_classifier_predictions = knn_classifier.predict(x_test)
```

```python
# calculating the accuracy and the confusion matrix
from sklearn.metrics import accuracy_score, confusion_matrix
acc = accuracy_score(knn_classifier_predictions, y_test)
cm = confusion_matrix(knn_classifier_predictions, y_test)
print(acc)
cm
```

```python
# getting the score of the training and testing dataset
print(knn_classifier.score(x_train, y_train))
print(knn_classifier.score(x_test, y_test))
```

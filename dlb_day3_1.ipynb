```python
# linear regression of wine quality 
# predicting quality of wine through linear regression model

import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/deep_learning_bootcamp/winequality-red.csv')
print('Wine dataframe looks like \n', dataset)
```

```python
# convert data into array
wine_dataset = np.asarray(dataset)
```

```python
wine_dataset
```

```python
# selecting input features 
wine_dataset_features = wine_dataset[:, 0:11]
print('wine dataset features value are: \n', wine_dataset_features)
print('shape of wine dataset is: ', np.shape(wine_dataset_features), '\n')
```

```python
# selecting target variable
wine_dataset_target = wine_dataset[:, 11]
print('wine dataset target values are: \n', wine_dataset_target)
print('shape of wine dataset target values are: ', np.shape(wine_dataset_target), "\n")
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(wine_dataset_features, wine_dataset_target, test_size=0.7)
```

```python
print('traning data is:\n', x_train, '\n')
print('shape of the training dataset is:\n', np.shape(x_train), '\n')
print('trainng labels are:\n', y_train, '\n')
print('shape of the training label is:\n', np.shape(y_train), '\n')
```

```python
from sklearn.linear_model import LinearRegression
linear_regression_model = LinearRegression(fit_intercept=True, normalize=True)
```

```python
print(linear_regression_model)
linear_regression_model.fit(x_train, y_train)
```

```python
# print the coefficient and the intercept
print("linear regression coefficient: ", linear_regression_model.coef_)
print('linear regression intercept: ', linear_regression_model.intercept_)
```

```python
# making prediction on the testing dataset
linear_regression_prediction = linear_regression_model.predict(x_test)
print('the prediction of the linear regressor are: \n', linear_regression_prediction)
```

```python
# analyze the performance of the linear regression
from sklearn.metrics import mean_absolute_error
mean_absolute_error(linear_regression_prediction, y_test)
```

```python
# calculate the mean sqaure error
from sklearn.metrics import mean_squared_error
mean_squared_error(linear_regression_prediction, y_test)
```

```python
# r2 score
print("comparing the predictions with the ground truth: \n", np.column_stack((linear_regression_prediction[:10], y_test[:10])), "\n\n")
r = r2_score(linear_regression_prediction, y_test)
r
```

```python
# linear regression on covid dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/deep_learning_bootcamp/allahabad.csv')
```

```python
dataset
```

```python
dataset.describe()
```

```python
x = dataset['Day'].values
y = dataset['Confirmed'].values
```

```python
print(x.shape)
print(y.shape)
```

```python
# reshape to 2D array
x = dataset['Day'].values.reshape(-1,1)
y = dataset['Confirmed'].values.reshape(-1,1)
```

```python
x.shape
y.shape
```

```python
Regressor = LinearRegression()
Regressor.fit(x, y)
```

```python
# prediction
y_pred = Regressor.predict(x)
```

```python
# using matplotlib
plt.scatter(x, y, color='magenta')
plt.plot(x, y_pred, color='green')
plt.title('daywise confirmed case in allahabad')
plt.xlabel('no_of_days')
plt.ylabel('confirmed_cases')
plt.show()
```

```python
# polynomial regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 5)
```

```python
x_poly = poly.fit_transform(x)
Regressor = LinearRegression()
Regressor.fit(x_poly, y)
```

```python
# prediction
y_poly_pred = Regressor.predict(x_poly)
```

```python
# using matplotlib
plt.scatter(x, y, color='yellow')
plt.plot(x, y_poly_pred, color='green')
plt.title('daywise confirmed case in allahabad')
plt.xlabel('no_of_days')
plt.ylabel('confirmed_cases')
plt.show()
```

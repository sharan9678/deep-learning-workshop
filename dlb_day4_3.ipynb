```python
import pandas as pd
import numpy as np
```

```python
# rounding off the values 
np.set_printoptions(suppress=True, precision=3)
```

```python
# loading the dataset
dataset = pd.read_csv('/content/drive/MyDrive/deep_learning_bootcamp/day4/iris.csv', header=None)
print(dataset)
```

```python
# transform the dataset into numpy metrices
iris_dataset = dataset.to_numpy()
print(iris_dataset)
print(iris_dataset.shape)
```

```python
# getting the features and labels from the dataset
iris_dataset_features = iris_dataset[1:, 0:4]
iris_dataset_labels = iris_dataset[1:, 4]
print(np.shape(iris_dataset_features))
print(np.shape(iris_dataset_labels))
```

```python
# split the dataset into training and testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris_dataset_features, iris_dataset_labels, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# normalize the training ans testing dataset
x_train_normalized = scaler.fit_transform(x_train)
x_test_normalised = scaler.transform(x_test)
```

```python
# defining our support vector machine(SVM) classifier
from sklearn.svm import SVC
svm_classifier = SVC(C=5.0, kernel="sigmoid")
print("SVM Classifier: \n", svm_classifier, "\n")
```

```python
# training our classifier
svm_classifier.fit(x_train_normalized, y_train)
```

```python
# getting the classifier elements
print(svm_classifier.support_vectors_)
print(np.shape(svm_classifier.support_vectors_)[0])
print(svm_classifier.n_support_)
print(svm_classifier.intercept_)
```

```python
# getting the prediction on the train data
svm_train_predictions = svm_classifier.predict(x_train_normalized)
print(svm_train_predictions)
```

```python
# comparing the prediction with the groud truth on training data
print(np.column_stack((svm_train_predictions, y_train)))
```

```python
svm_classifier.score(x_train_normalized, y_train)
```

```python
# getting the cofusion matrix
from sklearn.metrics import confusion_matrix
train_cm = confusion_matrix(y_train, svm_train_predictions)
print(train_cm)
```

```python
import seaborn as sn
sn.heatmap(train_cm, annot=True)
```

```python
# making prediction using 
test_accuracy = svm_classifier.score(x_test_normalised, y_test)
print(test_accuracy)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, svm_train_predictions))
```
